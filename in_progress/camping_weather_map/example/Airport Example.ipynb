```python
import altair as alt
import pandas as pd
```

```python
# Airport Example
```

```python
url = "https://raw.githubusercontent.com/deldersveld/topojson/master/countries/united-states/us-albers.json"
feature = "us"

source = alt.topo_feature(url, feature)
```

```python
df = pd.read_csv('D:/jadle/Documents/Python/Personal/Camping Map/Example/airports.csv')
df = df[df.type == 'large_airport']
df = df[pd.isna(df['continent'])]
df2 = df.groupby(by=['iso_country'])['name'].count().to_frame().reset_index()
df2.rename(columns={'name' : 'number_of_airports'},inplace=True)
```

```python
base = alt.Chart(source).mark_geoshape(
    fill='lightgray',
    stroke='white'
).project('albersUsa').properties(
    width=800,
    height=600
)


points = alt.Chart(df).mark_circle().encode(
    longitude='longitude_deg:Q',
    latitude='latitude_deg:Q',
    size=alt.Size('number_of_airports:Q', title='# airports'),
    color=alt.value('red')
).properties(
    width=500,
    height=400
).transform_lookup(
    lookup='iso_country',
    from_=alt.LookupData(df2, 'iso_country', ['iso_country', 'number_of_airports'])
)

base + points
```
